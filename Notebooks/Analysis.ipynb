```python
import matplotlib.pyplot as plt
import numpy as np

# Data: Accuracies for 5 classes and 3 methods
classes = ['KM', 'WKHM', 'CKM', 'KC', 'KHM']
methods = ['Basic Prompting', 'Contextualised Prompting', 'Fine Tuning']

# Example accuracy data for each method and class
accuracies = [
    [14.28, 10.17, 17.4,28.33,33.33],  # Method A
    [48.33, 56.66, 70.83, 77.5, 53.33],  # Method B
    [90.83, 100, 98.33, 93.33, 90],  # Method C
]

# Convert to numpy array for easier manipulation
accuracies = np.array(accuracies)

# Define bar width and positions
bar_width = 0.25
x_indices = np.arange(len(classes))

# Create the plot
plt.figure(figsize=(10, 6))

# Plot bars for each method
for i, method in enumerate(methods):
    plt.bar(x_indices + i * bar_width, accuracies[i], width=bar_width, label=method)

# Add labels, title, and legend
plt.xlabel('Classes', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.title('Accuracy Comparison by Class and Method', fontsize=14)
plt.xticks(x_indices + bar_width, classes)
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Calculate the average accuracy across all classes for each method
combined_accuracies = accuracies.mean(axis=1)

# Plot the combined accuracies
plt.figure(figsize=(8, 5))
plt.bar(methods, combined_accuracies, color=['blue', 'orange', 'green'])

# Add labels, title, and legend
plt.ylabel('Combined Accuracy (%)', fontsize=12)
plt.title('Combined Accuracy Across All Classes', fontsize=14)
plt.ylim(0, 100)  # Set y-limit to 0-100
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plot individual accuracies for each class
plt.figure(figsize=(10, 6))

for i, class_name in enumerate(classes):
    plt.plot(methods, accuracies[:, i], marker='o', label=class_name)

# Add labels, title, and legend
plt.xlabel('Methods', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.title('Accuracy Trends by Class', fontsize=14)
plt.legend(title='Classes', loc='best')
plt.tight_layout()

# Show plot
plt.show()
```

```python
import json
import re
content = data.strip()
content = content[1:len(content)-1]
json_string = content.strip()
json_objects = []
cleaned_string = json_string.strip()

# Initialize variables to accumulate the JSON objects
current_object = ""
brace_count = 0

# Loop through each character in the cleaned string
for char in cleaned_string:
    # Start of a JSON object
    if char == '{':
        brace_count += 1

    if brace_count > 0:
        current_object += char  # Accumulate characters for the current object

    # End of a JSON object
    if char == '}':
        brace_count -= 1
        if brace_count == 0:
            # Try parsing the current object
            try:
                # Remove any unnecessary newlines or spaces between objects (but keep within objects)
                clean_json_object = re.sub(r'\\n', "", current_object)
                #print(clean_json_object[2])
                json_objects.append(json.loads(clean_json_object))
                current_object = ""  # Reset for the next object
            except json.JSONDecodeError as e:
                print(f"Error parsing JSON: {e}")
                print(f"Problematic JSON: {current_object}")
                current_object = ""  # Reset on failure

# Output the result
print(f"Successfully parsed {len(json_objects)} JSON objects:")
#print(json.dumps(json_objects, indent=4))
print(json_objects[0])
```
